### (1) Ìñ•ÌõÑ 20ÎÖÑ Í≥†Î†πÌôî ÎπÑÏú®(%) Î≥ÄÌôî (60ÏÑ∏ Í∏∞Ï§Ä)  

```python
#Îç∞Ïù¥ÌÑ∞ ÏùΩÏñ¥Ïò§Í∏∞
pop=pd.read_excel("Ï†ÑÍµ≠_Ïù∏Íµ¨Ïàò.xlsx")

#ÌñâÍ≥º Ïó¥ Î∞îÍæ∏Í∏∞
pop=pop.T

#Ï≤´Î≤àÏ®∞ Ìñâ Ïª¨ÎüºÎ™ÖÏúºÎ°ú ÏßÄÏ†ï ÌõÑ Ï≤´Î≤àÏß∏ Ìñâ ÏÇ≠Ï†ú
pop = pop.rename(columns=pop.iloc[0])
pop = pop.drop(pop.index[0])
```

```python
#Î≥¥Í≥†Ïã∂ÏùÄ ÌñâÎßå Ï†ÄÏû•
pop=pop.loc[['ÌòÑÏû¨_Í≥†Î†πÏù∏Íµ¨ÎπÑÏú®','15ÎÖÑÌõÑ_Í≥†Î†πÏù∏Íµ¨ÎπÑÏú®','20ÎÖÑÌõÑ_Í≥†Î†πÏù∏Íµ¨ÎπÑÏú®'],:]
pop
```

- ÏãúÍ∞ÅÌôî

(1) ver1 : Í∞ÄÎ°úÏÑ† ÌëúÏãúx

```python
fig = plt.figure(figsize=(15,5)) ## Ï∫îÎ≤ÑÏä§ ÏÉùÏÑ±
fig.set_facecolor('white') ## Ï∫îÎ≤ÑÏä§ ÏÉâÏÉÅ ÏÑ§Ï†ï
ax = fig.add_subplot() ## Í∑∏Î¶º ÎºàÎåÄ(ÌîÑÎ†àÏûÑ) ÏÉùÏÑ±

year=['ÌòÑÏû¨', '15ÎÖÑÌõÑ', '20ÎÖÑÌõÑ']

#ax.plot(xÏ∂ï, yÏ∂ï, marker='o', linewidth=ÏÑ†ÍµµÍ∏∞ Ï°∞Ï†ïÍ∞ÄÎä•, color='ÏÉâÏßÄÏ†ï')
ax.plot(year,pop['Ï†ÑÍµ≠'], marker='o', linewidth=3, color='limegreen') 

plt.title('Ìñ•ÌõÑ Í≥†Î†πÏù∏Íµ¨ÎπÑÏú® Î≥ÄÌôî(%) (60ÏÑ∏ Í∏∞Ï§Ä)',fontsize=13) ## ÌÉÄÏù¥ÌãÄ ÏÑ§Ï†ï
plt.show()
```

(2) ver2 : 50% Î∂ÄÎ∂ÑÏóê Í∞ÄÎ°úÏÑ† ÌëúÏãúo  
-> Í∑∏ÎûòÌîÑÏùò Ïù¥Ìï¥Î•º ÎèïÍ∏∞ÏúÑÌï¥ 50% ÏßÄÏ†êÏóê ÏàòÌèâÏÑ†ÏùÑ Ï∂îÍ∞ÄÌï¥Î≥¥ÏïòÏäµÎãàÎã§!

```python
fig = plt.figure(figsize=(15,5)) ## Ï∫îÎ≤ÑÏä§ ÏÉùÏÑ±
fig.set_facecolor('white') ## Ï∫îÎ≤ÑÏä§ ÏÉâÏÉÅ ÏÑ§Ï†ï
ax = fig.add_subplot() ## Í∑∏Î¶º ÎºàÎåÄ(ÌîÑÎ†àÏûÑ) ÏÉùÏÑ±

year=['ÌòÑÏû¨', '15ÎÖÑÌõÑ', '20ÎÖÑÌõÑ']

#50% ÏßÄÏ†êÏóê Í∞ÄÎ°ú ÏÑ† ÌëúÏãú
plt.axhline(50.0, 0.0, 1.0, color='lightgray', linestyle='--', linewidth=2)

#ax.plot(xÏ∂ï, yÏ∂ï, marker='o', linewidth=ÏÑ†ÍµµÍ∏∞ Ï°∞Ï†ïÍ∞ÄÎä•, color='ÏÉâÏßÄÏ†ï')
ax.plot(year,pop['Ï†ÑÍµ≠'], marker='o', linewidth=3, color='limegreen') 

plt.title('Ìñ•ÌõÑ Í≥†Î†πÏù∏Íµ¨ÎπÑÏú® Î≥ÄÌôî(%) (60ÏÑ∏ Í∏∞Ï§Ä)',fontsize=13) ## ÌÉÄÏù¥ÌãÄ ÏÑ§Ï†ï
plt.show()
```

### (2) ÎØ∏Îûò Î≥µÏßÄÏÑºÌÑ∞ Í∞úÏàò (60ÏÑ∏ Í∏∞Ï§Ä)  
-> 50ÏÑ∏~60ÏÑ∏Îäî ÏïÑÏßÅ ÏÉùÏÇ∞ÌôúÎèôÏùÑ ÌïòÎäî Ïù∏Íµ¨ÎùºÍ≥† ÏÉùÍ∞ÅÌï¥ÏÑú 60ÏÑ∏Î•º Í∏∞Ï§ÄÏúºÎ°ú Í∑∏ÎûòÌîÑÎ•º ÏãúÍ∞ÅÌôîÌïòÏòÄÏäµÎãàÎã§!  

-> Í≥ÑÏÇ∞Î∞©Î≤ïÏùÄ "Ï∂îÍ∞ÄÎ≥µÏßÄÏÑºÌÑ∞_Î∂ÑÏÑù" Ìè¥ÎçîÏùò "Ï∂îÍ∞ÄÎ≥µÏßÄÏÑºÌÑ∞_Í≥ÑÏÇ∞Î∞©Î≤ï.xlsx"ÏùÑ Ï∞∏Í≥†Ìï¥Ï£ºÏÑ∏Ïöî!

```python
#Îç∞Ïù¥ÌÑ∞ ÏùΩÏñ¥Ïò§Í∏∞
center=pd.read_excel("Ï∂îÍ∞ÄÎ≥µÏßÄÏÑºÌÑ∞_Ïà´ÏûêÎßå.xlsx")

#ÌñâÍ≥ºÏó¥ Î∞îÍæ∏Í∏∞
center=center.T

#Ï≤´Î≤àÏ®∞ Ìñâ Ïª¨ÎüºÎ™ÖÏúºÎ°ú ÏßÄÏ†ï ÌõÑ Ï≤´Î≤àÏß∏ Ìñâ ÏÇ≠Ï†ú
center = center.rename(columns=center.iloc[0])
center = center.drop(center.index[0])
```

```python
#Î≥¥Í≥†Ïã∂ÏùÄ Ïª¨ÎüºÎßå Ï†ÄÏû•
center=center[["ÏöîÏñëÏãúÏÑ§Í∞úÏàò"]]
center
```

```python
#15ÎÖÑÌõÑ, 20ÎÖÑÌõÑÎßå Ï†ÄÏû•
center=center.loc[['ÌòÑÏû¨','15ÎÖÑÌõÑ','20ÎÖÑÌõÑ'],:]
center
```

- ÏãúÍ∞ÅÌôî

```python
fig = plt.figure(figsize=(15,5)) ## Ï∫îÎ≤ÑÏä§ ÏÉùÏÑ±
fig.set_facecolor('white') ## Ï∫îÎ≤ÑÏä§ ÏÉâÏÉÅ ÏÑ§Ï†ï
ax = fig.add_subplot() ## Í∑∏Î¶º ÎºàÎåÄ(ÌîÑÎ†àÏûÑ) ÏÉùÏÑ±

year=['ÌòÑÏû¨', '15ÎÖÑÌõÑ', '20ÎÖÑÌõÑ']
#ax.plot(xÏ∂ï, yÏ∂ï, marker='o', linewidth=ÏÑ†ÍµµÍ∏∞ Ï°∞Ï†ïÍ∞ÄÎä•)
ax.plot(year,center['ÏöîÏñëÏãúÏÑ§Í∞úÏàò'], marker='o', linewidth=3, color='limegreen') 
 
plt.title('ÎØ∏Îûò Î≥µÏßÄÏÑºÌÑ∞ Í∞úÏàò',fontsize=13) ## ÌÉÄÏù¥ÌãÄ ÏÑ§Ï†ï
plt.show()
```
