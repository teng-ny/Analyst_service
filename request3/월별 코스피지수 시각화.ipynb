### (1) 5Í∞úÎÖÑ(18-22ÎÖÑ) ÏõîÎ≥Ñ ÏΩîÏä§ÌîºÏßÄÏàò ÏãúÍ∞ÅÌôî

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Í≤ΩÍ≥†Î¨¥Ïãú
import warnings
warnings.filterwarnings('ignore')
# ÌïúÍ∏ÄÏ≤òÎ¶¨
plt.rc('font', family='Malgun Gothic')
# '-' Î¨∏Ïûê Ï≤òÎ¶¨
plt.rc('axes', unicode_minus=False)
```

```python
#Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
kospi=pd.read_csv("result_month.csv", encoding='euc-kr')


#Î∂ÑÏÑùÏóê ÌïÑÏöîÌïú Ïª¨ÎüºÎßå Ï∂îÏ∂ú
kospi=kospi[['ÎÇ†Ïßú', 'Ï¢ÖÍ∞Ä']]
#Ïª¨ÎüºÎ™Ö Î≥ÄÍ≤Ω
kospi.columns=['ÎÇ†Ïßú', 'ÏΩîÏä§ÌîºÏßÄÏàò']


#Îç∞Ïù¥ÌÑ∞ÌÉÄÏûÖ datetimeÏúºÎ°ú Î≥ÄÍ≤Ω
import datetime
kospi['ÎÇ†Ïßú']=pd.to_datetime(kospi['ÎÇ†Ïßú'])


#ÎÇ†ÏßúÏóêÏÑú ÎÖÑÎèÑÏôÄ ÏõîÏùÑ Î∂ÑÎ¶¨
kospi['year'] = kospi['ÎÇ†Ïßú'].dt.strftime("%Y")
kospi['month'] = kospi['ÎÇ†Ïßú'].dt.strftime("%m")


#ÎÇ†Ïßú Ïª¨Îüº "ÎÖÑÎèÑ.Ïõî" ÌòïÏãùÏúºÎ°ú Î≥ÄÍ≤Ω
kospi['ÎÇ†Ïßú']=kospi['year'] + '.' + kospi['month']


#ÏãúÍ∞ÅÌôî ÏãúÏóê Í∞ÄÎèÖÏÑ±ÏùÑ ÏúÑÌï¥ 2018.01->18.02 ÌòïÏãùÏúºÎ°ú Î≥ÄÍ≤Ω
kospi['ÎÇ†Ïßú']=kospi['ÎÇ†Ïßú'].str[2:]
```

-> Î≥ÄÍ≤ΩÎêú Í≤ÉÏùÑ ÌôïÏù∏

- (1) ÎßàÏª§ ÌëúÏãúver

```python
#xÏ∂ïÏúºÎ°ú ÏßÄÏ†ïÌï† Ïª¨Îüº 'month'Ïóê Ï†ÄÏû•
month=kospi['ÎÇ†Ïßú']

fig = plt.figure(figsize=(30,8)) ## Ï∫îÎ≤ÑÏä§ ÏÉùÏÑ±
fig.set_facecolor('white') ## Ï∫îÎ≤ÑÏä§ ÏÉâÏÉÅ ÏÑ§Ï†ï
ax = fig.add_subplot() ## Í∑∏Î¶º ÎºàÎåÄ(ÌîÑÎ†àÏûÑ) ÏÉùÏÑ±

#ax.plot(xÏ∂ï, yÏ∂ï, marker='o', linewidth=ÏÑ†ÍµµÍ∏∞ Ï°∞Ï†ïÍ∞ÄÎä•)
ax.plot(month,kospi['ÏΩîÏä§ÌîºÏßÄÏàò'], marker='o', linewidth=3) 
 
plt.title('18-22ÎÖÑÎèÑ ÏõîÎ≥Ñ ÏΩîÏä§ÌîºÏßÄÏàò ÏãúÍ∞ÅÌôî',fontsize=20) ## ÌÉÄÏù¥ÌãÄ ÏÑ§Ï†ï
plt.show()
```
